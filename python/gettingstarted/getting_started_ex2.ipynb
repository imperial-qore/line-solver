```python
from line_solver import *
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```python
model = Network('M/G/1')
source = Source(model, 'Source')
queue = Queue(model, 'Queue', SchedStrategy.FCFS)
sink = Sink(model, 'Sink')
jobclass1 = OpenClass(model, 'Class1')
jobclass2 = OpenClass(model, 'Class2')
source.setArrival(jobclass1, Exp(0.5))
source.setArrival(jobclass2, Exp(0.5))
queue.setService(jobclass1, Erlang.fitMeanAndSCV(1, 1 / 3))
queue.setService(jobclass2, Replayer.fitAPH("example_trace.txt"))
```

```python
P = model.initRoutingMatrix()
P.set(jobclass1, jobclass1, source, queue, 1.0)
P.set(jobclass1, jobclass1, queue, sink, 1.0)
P.set(jobclass2, jobclass2, source, queue, 1.0)
P.set(jobclass2, jobclass2, queue, sink, 1.0)
model.link(P)
```

```python
SolverJMT(model,'seed',23000).getAvgTable()
```
