# Gallery Example: APH/M/1 Queue

This example demonstrates an APH/M/1 queueing system:
- **Arrivals**: Acyclic Phase-type (APH) distribution
- **Service**: Exponential service times
- **Servers**: 1 server
- **Capacity**: Infinite
- **Scheduling**: FCFS

APH distributions are phase-type distributions with an acyclic structure, providing flexibility in modeling arrival processes with specified moments.

```python
from line_solver import *
GlobalConstants.set_verbose(VerboseLevel.STD)
```

```python
def gallery_aphm1():
    """Create APH/M/1 queueing model"""
    model = Network('APH/M/1')
    
    # Block 1: nodes
    source = Source(model, 'mySource')
    queue = Queue(model, 'myQueue', SchedStrategy.FCFS)
    sink = Sink(model, 'mySink')
    
    # Block 2: classes
    oclass = OpenClass(model, 'myClass')
    # APH distribution fitted to mean=1, variance=0.99, skewness=1.999
    source.set_arrival(oclass, APH.fit_central(1, 0.99, 1.999))
    queue.set_service(oclass, Exp(2))
    
    # Block 3: topology
    model.link(Network.serial_routing(source, queue, sink))
    
    return model

# Create the model
model = gallery_aphm1()
```

## About APH Distributions

APH (Acyclic Phase-type) distributions:
- Can be fitted to match specified moments (mean, variance, skewness)
- Have an acyclic structure making them computationally efficient

- Suitable for modeling complex arrival/service processes

```python
# Solve with multiple solvers
print("\n=== Solver Results ===")

# CTMC Solver
solver_ctmc = CTMC(model, cutoff=15)
avg_table_ctmc = solver_ctmc.avg_table()
print("\nCTMC Solver:")
print(avg_table_ctmc)

# Fluid Solver
solver_fluid = FLD(model)
avg_table_fluid = solver_fluid.avg_table()
print("\nFluid Solver:")
print(avg_table_fluid)
```
