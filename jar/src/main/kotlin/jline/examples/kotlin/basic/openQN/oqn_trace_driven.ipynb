```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.jmt.*
import jline.util.matrix.*
import java.io.File
```

```kotlin
// Create the model
val model = Network("model")

// Define network nodes
val source = Source(model, "Source")
val queue = Queue(model, "Queue", SchedStrategy.FCFS)
val sink = Sink(model, "Sink")

// Define job class
val jobclass = OpenClass(model, "OpenClass", 0)
```

```kotlin
// Set arrival process
source.setArrival(jobclass, Exp(1.0))
```

```kotlin
// Set service process using trace file
// The trace file should be in the same directory as the MATLAB example
val traceFilePath = "/home/gcasale/line-dev.git/matlab/examples/basic/openQN/example_trace.txt"

// Set service process using Replayer with the trace file
queue.setService(jobclass, Replayer(traceFilePath))
```

```kotlin
// Define network topology using serial routing
model.link(Network.serialRouting(source, queue, sink))
```

```kotlin
// Solve using JMT solver
val avgTable = JMT(model, "seed", 23000).avgTable
avgTable.print()
```
