```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.mva.*
import jline.util.matrix.*
```

```kotlin
// Service times - S(i,r) - mean service time of class r at station i
val D = Matrix.of(arrayOf(
    doubleArrayOf(10.0, 5.0),
    doubleArrayOf(5.0, 9.0)
))

// Number of jobs - N(r) - number of jobs of class r
val N = Matrix.of(doubleArrayOf(1.0, 2.0))

// Think times - Z(r) - mean service time of class r at delay station i
val Z = Matrix.of(arrayOf(
    doubleArrayOf(91.0, 92.0),
    doubleArrayOf(93.0, 94.0)
))
```

```kotlin
// Create cyclic model using the Network.cyclicPsInf method
val model = Network.cyclicPsInf(N, D, Z)
```

```kotlin
// Solve using exact MVA
val avgTable = MVA(model, "method", "exact").getAvgTable()
avgTable.print()
```

The results show performance metrics for each station and job class:
- **Station**: Queue1 and Queue2
- **JobClass**: Class1 and Class2
- **QLen**: Average queue length
- **Util**: Utilization
- **RespT**: Response time
- **ResidT**: Residence time
- **ArvR**: Arrival rate
- **Tput**: Throughput
