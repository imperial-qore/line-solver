# Gallery Example: Er/M/1-PS Queue

This example demonstrates an Erlang/M/1-PS queueing system:
- **Arrivals**: Erlang distribution (less variable than Poisson)
- **Service**: Exponential service times
- **Servers**: 1 server
- **Capacity**: Infinite
- **Scheduling**: Processor Sharing (PS)

In processor sharing, all jobs in the system share the server capacity equally, which is different from FCFS scheduling.

```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.ctmc.*
import jline.solvers.fluid.*
import jline.solvers.mva.*
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```kotlin
fun gallery_erlm1ps(): Network {
    """Create Er/M/1-PS queueing model"""
    val model = Network("Er/M/1-PS")
// Block 1: nodes
    val source = Source(model, "mySource")
    val queue = Queue(model, "myQueue", SchedStrategy.PS)  # Processor Sharing
    val sink = Sink(model, "mySink")
// Block 2: classes
    val oclass = OpenClass(model, "myClass")
// Erlang arrival with mean=1 and order=5 (less variable than exponential)
    source.setArrival(oclass, Erlang.fit_mean_and_order(1, 5))
// Exponential service with rate=2 (mean service time = 0.5)
    queue.setService(oclass, Exp(2))
// Block 3: topology
    model.link(Network.serialRouting(source, queue, sink))
    
    return model
// Create the model
val model = gallery_erlm1ps()
```

## About Processor Sharing

In processor sharing scheduling:
- All jobs receive service simultaneously with equal share of capacity
- Response time is independent of service order
- Useful for modeling time-shared computer systems
- Provides fairness among jobs

```kotlin
// Solve with multiple solvers
println("\n=== Solver Results ===")
// MVA Solver
val solver_mva = MVA(model)
val avg_table_mva = solver_mva.avgTable
println("\nMVA Solver:")
println(avg_table_mva)
// CTMC Solver
val solver_ctmc = CTMC(model, "cutoff", 15)
val avg_table_ctmc = solver_ctmc.avgTable
println("\nCTMC Solver:")
println(avg_table_ctmc)
// Fluid Solver
val solver_fluid = FLD(model)
val avg_table_fluid = solver_fluid.avgTable
println("\nFluid Solver:")
println(avg_table_fluid)
```
