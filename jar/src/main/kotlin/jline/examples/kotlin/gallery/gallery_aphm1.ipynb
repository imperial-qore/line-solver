# Gallery Example: APH/M/1 Queue

This example demonstrates an APH/M/1 queueing system:
- **Arrivals**: Acyclic Phase-type (APH) distribution
- **Service**: Exponential service times
- **Servers**: 1 server
- **Capacity**: Infinite
- **Scheduling**: FCFS

APH distributions are phase-type distributions with an acyclic structure, providing flexibility in modeling arrival processes with specified moments.

```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.ctmc.*
import jline.solvers.fluid.*
import jline.solvers.mva.*
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```kotlin
fun gallery_aphm1(): Network {
    """Create APH/M/1 queueing model"""
    val model = Network("APH/M/1")
// Block 1: nodes
    val source = Source(model, "mySource")
    val queue = Queue(model, "myQueue", SchedStrategy.FCFS)
    val sink = Sink(model, "mySink")
// Block 2: classes
    val oclass = OpenClass(model, "myClass")
// APH distribution fitted to mean=1, variance=0.99, skewness=1.999
    source.setArrival(oclass, APH.fit_central(1, 0.99, 1.999))
    queue.setService(oclass, Exp(2))
// Block 3: topology
    model.link(Network.serialRouting(source, queue, sink))
    
    return model
// Create the model
val model = gallery_aphm1()
```

## About APH Distributions

APH (Acyclic Phase-type) distributions:
- Can be fitted to match specified moments (mean, variance, skewness)
- Have an acyclic structure making them computationally efficient
- Are dense in the space of distributions on [0, ∞)
- Suitable for modeling complex arrival/service processes

```kotlin
// Solve with multiple solvers
println("\n=== Solver Results ===")
// CTMC Solver
val solver_ctmc = CTMC(model, "cutoff", 15)
val avg_table_ctmc = solver_ctmc.avgTable
println("\nCTMC Solver:")
println(avg_table_ctmc)
// Fluid Solver
val solver_fluid = FLD(model)
val avg_table_fluid = solver_fluid.avgTable
println("\nFluid Solver:")
println(avg_table_fluid)
```
