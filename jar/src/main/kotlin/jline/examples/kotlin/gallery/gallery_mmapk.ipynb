# Gallery Example: M/MAP/k Queue

This example demonstrates an M/MAP/k queueing system:
- **Arrivals**: Exponential (Poisson process)
- **Service**: Markovian Arrival Process (MAP)
- **Servers**: k servers (default k=2)
- **Capacity**: Infinite
- **Scheduling**: FCFS

This extends the M/MAP/1 queue to multiple servers.

```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.ctmc.*
import jline.solvers.fluid.*
import jline.solvers.mva.*
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```kotlin
fun gallery_mmapk(map_dist=None, k=2): Network {
    """Create M/MAP/k queueing model"""
    if map_dist is None:
        val map_dist = MAP.rand()
    
    val model = Network("M/MAP/k")
// Block 1: nodes
    val source = Source(model, "mySource")
    val queue = Queue(model, "myQueue", SchedStrategy.FCFS)
    val sink = Sink(model, "mySink")
// Block 2: classes
    val oclass = OpenClass(model, "myClass")
    source.setArrival(oclass, Exp(1))  # Poisson arrivals with rate 1
    queue.setService(oclass, map_dist)  # MAP service
    queue.setNumberOfServers(k)
// Block 3: topology
    model.link(Network.serialRouting(source, queue, sink))
    
    return model
// Create the model with k=2 servers
val model = gallery_mmapk(k=2)
```

## About M/MAP/k Queues

The M/MAP/k queue combines:
- **Poisson arrivals**: Simple, uncorrelated arrival process
- **MAP service**: Complex service time distribution with potential correlation
- **Multiple servers**: Parallel service capacity

This model is useful when:
- Service times have phases or modes
- Service completion rates depend on server state
- You need to model complex service behavior with parallel capacity

```kotlin
// Solve with multiple solvers
println("\n=== Solver Results ===")
// CTMC Solver
val solver_ctmc = CTMC(model, "cutoff", 15)
val avg_table_ctmc = solver_ctmc.avgTable
println("\nCTMC Solver:")
println(avg_table_ctmc)
// Fluid Solver
val solver_fluid = FLD(model)
val avg_table_fluid = solver_fluid.avgTable
println("\nFluid Solver:")
println(avg_table_fluid)
```
