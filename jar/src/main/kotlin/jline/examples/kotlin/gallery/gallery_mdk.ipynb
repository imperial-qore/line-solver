# Gallery Example: M/D/k Queue

This example demonstrates an M/D/k queueing system:
- **Arrivals**: Exponential (Poisson process)
- **Service**: Deterministic (constant service times)
- **Servers**: k servers (default k=2)
- **Capacity**: Infinite
- **Scheduling**: FCFS

The deterministic service combined with multiple servers provides very predictable processing.

```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.ctmc.*
import jline.solvers.fluid.*
import jline.solvers.mva.*
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```kotlin
fun gallery_mdk(k=2): Network {
    """Create M/D/k queueing model"""
    val model = Network("M/D/k")
// Block 1: nodes
    val source = Source(model, "mySource")
    val queue = Queue(model, "myQueue", SchedStrategy.FCFS)
    val sink = Sink(model, "mySink")
// Block 2: classes
    val oclass = OpenClass(model, "myClass")
// Exponential arrivals with mean=1
    source.setArrival(oclass, Exp.fitMean(1))
// Deterministic service with mean=2/k (so total capacity matches arrival rate)
    queue.setService(oclass, Det(2/k))
    queue.setNumberOfServers(k)
// Block 3: topology
    model.link(Network.serialRouting(source, queue, sink))
    
    return model
// Create the model with k=2 servers
val model = gallery_mdk(k=2)
```

## About M/D/k Queues

For M/D/k with k=2 servers:
- **Arrival rate**: λ = 1 (mean inter-arrival = 1)
- **Service time**: D = 2/k = 1.0 (deterministic)
- **Service rate per server**: μ = 1
- **Total capacity**: k × μ = 2
- **Utilization**: ρ = λ / (k × μ) = 1/2 = 0.5

The deterministic service (SCV = 0) provides the best possible performance for given utilization, as there's no service variability to cause queue buildup.

```kotlin
// Solve with multiple solvers
println("\n=== Solver Results ===")
// CTMC Solver
val solver_ctmc = CTMC(model, "cutoff", 15)
val avg_table_ctmc = solver_ctmc.avgTable
println("\nCTMC Solver:")
println(avg_table_ctmc)
// Fluid Solver
val solver_fluid = FLD(model)
val avg_table_fluid = solver_fluid.avgTable
println("\nFluid Solver:")
println(avg_table_fluid)
```
