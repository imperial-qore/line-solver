# Gallery Example: Cox/M/1 Queue

This example demonstrates a Cox/M/1 queueing system:
- **Arrivals**: Coxian distribution
- **Service**: Exponential service times
- **Servers**: 1 server
- **Capacity**: Infinite
- **Scheduling**: FCFS

Coxian distributions are a special case of phase-type distributions with a series structure, commonly used for modeling service times with controlled variance.

```kotlin
// Kotlin notebook
import jline.*
import jline.lang.*
import jline.lang.nodes.*
import jline.lang.processes.*
import jline.lang.constant.*
import jline.solvers.ctmc.*
import jline.solvers.fluid.*
import jline.solvers.mva.*
GlobalConstants.setVerbose(VerboseLevel.STD)
```

```kotlin
fun gallery_coxm1(): Network {
    """Create Cox/M/1 queueing model"""
    val model = Network("Cox/M/1")
// Block 1: nodes
    val source = Source(model, "mySource")
    val queue = Queue(model, "myQueue", SchedStrategy.FCFS)
    val sink = Sink(model, "mySink")
// Block 2: classes
    val oclass = OpenClass(model, "myClass")
// Coxian distribution fitted to mean=1, variance=0.99, skewness=1.999
    source.setArrival(oclass, Coxian.fit_central(1, 0.99, 1.999))
    queue.setService(oclass, Exp(2))
// Block 3: topology
    model.link(Network.serialRouting(source, queue, sink))
    
    return model
// Create the model
val model = gallery_coxm1()
```

## About Coxian Distributions

Coxian distributions:
- Have a series structure where phases are visited sequentially
- Allow early exit from each phase with some probability
- Can model a wide range of variability (SCV from 0 to infinity)
- Are often used for service time modeling in queueing systems

```kotlin
// Solve with multiple solvers
println("\n=== Solver Results ===")
// CTMC Solver
val solver_ctmc = CTMC(model, "cutoff", 15)
val avg_table_ctmc = solver_ctmc.avgTable
println("\nCTMC Solver:")
println(avg_table_ctmc)
// Fluid Solver
val solver_fluid = FLD(model)
val avg_table_fluid = solver_fluid.avgTable
println("\nFluid Solver:")
println(avg_table_fluid)
```
